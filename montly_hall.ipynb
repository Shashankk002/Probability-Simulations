```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def simulate_monty_hall(n_trials, switch):
    wins = 0

    for _ in range(n_trials):
        doors = [0, 1, 2]
        car_position = np.random.choice(doors)
        player_choice = np.random.choice(doors)

        possible_doors_to_open = [door for door in doors if door != player_choice and door != car_position]
        monty_opens = np.random.choice(possible_doors_to_open)

        remaining_door = [door for door in doors if door != player_choice and door != monty_opens][0]

        final_choice = remaining_door if switch else player_choice

        if final_choice == car_position:
            wins += 1

    return wins / n_trials
```

```python
n_trials = 10000

stick_win_rate = simulate_monty_hall(n_trials, switch=False)
switch_win_rate = simulate_monty_hall(n_trials, switch=True)

print(f"Stick Strategy Win Rate:  {stick_win_rate * 100:.2f}%")
print(f"Switch Strategy Win Rate: {switch_win_rate * 100:.2f}%")
```

```python
strategies = ['Stick', 'Switch']
win_rates = [stick_win_rate, switch_win_rate]

plt.bar(strategies, win_rates, color=['red', 'green'])
plt.title("Monty Hall Strategy Win Rates")
plt.ylabel("Winning Probability")
plt.ylim(0, 1)
for i, rate in enumerate(win_rates):
    plt.text(i, rate + 0.02, f"{rate*100:.1f}%", ha='center', fontsize=12)
plt.grid(axis='y')
plt.show()
```
